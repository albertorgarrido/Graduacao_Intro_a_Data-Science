### Tipo Conjunto (sets):
Os itens são postos entre chaves e separados por vírgula.

- Set são sequências **não-ordenadas** e **não-indexados** de itens únicos. 
- Itens duplicados não são considerados

```python
numero_pares = {2,2,4,4,4,6,8,8,10,4}
```

```python
numero_pares
```

```python
numero_impares = {1,1,1,3,3,5,5,7,9}
```

```python
numero_impares
```

```python
numero_primos = {2,3,5,7}
```

```python
numero_primos
```

```python
numero_pares.difference(numero_primos)
```

```python
numero_impares.intersection(numero_primos)
```

### Tipo Dicionário:
Dicionários são sequências não ordenadas de itens na forma de pares key:values que se encontram dentro de chaves e separados por vírgula. Como value podemos usar objetos de qualquer tipo, já como key só podemos usar objetos de tipos imutáveis (strings, números ou tuplas).

Comandos interessantes: clear, copy, fromkeys, get, items, keys, pop, popitem, setdefault, update, values

```python
dicionario = {'chave1':[1,2,3],'chave2':2,'outra chave':4,45:26, 'outro':23}
dicionario
```

```python
dicionario.keys()
```

```python
dicionario.values()
```

```python
ficha = {'Peso':73, 'Altura':183, 'Idade':44, 'Nome':'Renato'}
```

```python
ficha['Idade']
```

```python
ficha.update({'Profissão':'Professor'})
```

```python
ficha.keys()
```

### Tipo Tupla:
Tuplas, assim como listas, são sequências ordenadas de itens. As diferenças entre elas são: os elementos estão entre parênteses e as tuplas são imutáveis, seus elementos não podem ser modificados.

```python
t = (1,2,3,4,7,4,3,1,'string',{1,2}, [1,2])
```

```python
type(t)
```

```python
t[0]
```

```python
t[5]
```

```python
t[3:6]
```

```python
lista = [1,2,3]
```

```python
tupla = tuple(lista)
```

```python
tupla
```

```python
lista2 = list(tupla)
```

```python
lista2
```

```python
d5 = {(1,2):'tupla'}
d5
```

```python
t2 = (2,4,7,2,2,2,2)
```

```python
t2.count(2)
```

```python
a = [1,2]
```

```python
t = (a,a,a,a,a)
l = [a,a,a,a,a]
```

```python
tupla_falsa = (1)
tupla_verdadeira = (1,)
print(type(tupla_falsa))
print(type(tupla_verdadeira))
```

Acessando cada elemento da tupla

```python
x, y = 10, 19 #Relembrando como declarar as variáveis em uma linha
print(y)
print(x)
```

Empacotamento e desempacotamento de tuplas

```python
x, y = 10, 19 #Relembrando como declarar as variáveis em uma linha
print(y)
print(x)
```

```python
t = (2,3,4)
x,y,z = t
print(t)
print(x)
print(y)
print(z)
```

```python
x, *y, z = (1,2,3,4)
print(x)
print(y)
print(z)
```

```python
# Troca de valores
print(x)
print(y)
x,y = y,x
print(x)
print(y)
```

## Iteráveis

Listas, dicionários, conjuntos e tuplas são objetos **iteráveis (iterable)**. Um iterável é qualquer objeto que é capaz de retornar seus elementos um a um, permitindo uma iteração via *for-loop*

Desempacotando iteráveis

```python
x,y = [1,2]
```

```python
z,w = (10,20)
```

```python
a,b = {'a':1,'b':2}
```
