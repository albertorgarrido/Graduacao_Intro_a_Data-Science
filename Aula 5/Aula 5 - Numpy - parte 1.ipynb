# Regressão Linear - Forma clássica

### $y = \alpha + \beta_1 x_1 + \epsilon$

### $\alpha = 4, \beta=3$

```python
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)
```

```python
plt.plot(X, y, "b.")
plt.xlabel("$x_1$", fontsize=18)
plt.ylabel("$y$", rotation=0, fontsize=18)
plt.axis([0, 2, 0, 15])
##save_fig("generated_data_plot")
plt.show()
```

### Forma fechada

### $(\mathbf{X^TX})^{-1} \mathbf{X^{T}}y$

```python
X_b = np.c_[np.ones((100, 1)), X]  # add x0 = 1 to each instance (multiplicar a constante)
theta_best = np.linalg.inv(X_b.T.dot(X_b)).dot(X_b.T).dot(y)
```

```python
theta_best
```

### Resultado $\hat{\alpha} = 4.21, \hat{\beta_1} = 2.77$

```python
X_new = np.array([[0], [2]])
X_new_b = np.c_[np.ones((2, 1)), X_new]  # add x0 = 1 to each instance
y_predict = X_new_b.dot(theta_best)
y_predict
```

Prevendo com novos dados

$\hat{y} = 4.21 + 2.77\times0$
<br>
$\hat{y} = 4.21 + 2.77\times2$

```python
plt.plot(X_new, y_predict, "r-")
plt.plot(X, y, "b.")
plt.axis([0, 2, 0, 15])
plt.show()
```
